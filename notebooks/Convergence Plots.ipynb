```python
from mms_pressure import run_mms_convergence
hvec, E = run_mms_convergence(8)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path


plt.rcParams.update({'font.size': 12})

savepath = (Path().parent / "../results/mms").resolve()
savepath.mkdir(exist_ok=True)

hvec = np.array(hvec)
plt.figure()
plt.loglog(hvec, E, "o-", markeredgecolor="k")
plt.xlabel("$h$", fontsize=16)
plt.ylabel("$||p - p_h||_{V}$", fontsize=16)

xlim, ylim = plt.gca().get_xlim(), plt.gca().get_ylim()
plt.autoscale(False)
plt.loglog([2e-2, 2e-1], [9e-3, 9e-1], "k--", lw=0.5, label="Quadratic")
plt.xlim(xlim)
plt.ylim(ylim)
# plt.legend()
plt.savefig(savepath / "pressure-convergence.png")
plt.show()
```

```python
from mms_solute import run_mms_solute_convergence
hvec, E = run_mms_solute_convergence(8)

plt.figure()
plt.loglog(hvec, E, "-o", markeredgecolor="k")
xlim, ylim = plt.gca().get_xlim(), plt.gca().get_ylim()
plt.autoscale(False)
plt.loglog([2e-2, 2e-1], [1e-1, 1e-0], "k--", lw=0.5, label="Linear")
plt.ylim(ylim)
plt.xlabel("$h$", fontsize=16)
plt.ylabel(r"$||c - c_h||_{L^2([0, T], V)}$", fontsize=16)
# plt.legend()
plt.savefig(savepath / "concentration-convergence.png")
plt.show()
```
